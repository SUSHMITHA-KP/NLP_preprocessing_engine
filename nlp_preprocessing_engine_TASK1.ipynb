```python
# 1. Difference between "Love" and "love" in NLP
# In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
# This means the model may interpret them as separate words, which can reduce consistency and
# affect learning. Converting all text to lowercase ensures both are treated as the same word,
# improving accuracy and reducing vocabulary size.

# 2. What happens if stopwords are not removed?
# If stopwords are not removed, the dataset will contain many common words like "is", "the", "and"
# that do not add much meaning. This increases noise, slows down processing, and can reduce model 
# performance because important words get less attention.

# 3. Two real-world scenarios where removing stopwords is harmful
# Sentiment Analysis: Words like "not" are critical.
# Example: "I am not happy" → removing "not" changes meaning completely.
# Question Answering Systems: Words like "what", "where", "how" are essential for understanding intent.

# 4. Difference between Stemming and Lemmatization
# Stemming: Cuts words to root form using simple rules (may not be meaningful)
# Example: "running" → "run", "better" → "bet"
# Lemmatization: Converts words to meaningful base form using vocabulary and grammar
# Example: "running" → "run", "better" → "good"
# Lemmatization is more accurate but slower than stemming.
```

```python
import re

def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""
    
    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r'\S+@\S+', "", text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)
    
    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    tokens = text.split()
    
    # Remove short words except meaningful ones
    meaningful_exceptions = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in meaningful_exceptions]
    
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)
    
    print("\n---------------------------")
    print("Original Text:", sentence)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    
    results.append((sentence, tokens, clean_sentence))
```

```python
def token_analysis(tokens):
    if len(tokens) == 0:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total
    
    return total, unique, avg_len

print("\n===== TOKEN ANALYSIS =====")

for sentence, tokens, _ in results:
    total, unique, avg_len = token_analysis(tokens)
    
    print("\nSentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
```

```python
from collections import Counter

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("\nTop 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(test_sentences)

print("\n===== FULL PIPELINE OUTPUT =====")
print(pipeline_output)
```

```python
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

print("\n===== ERROR HANDLING =====")

for case in edge_cases:
    tokens, clean_sentence = preprocess_text(case)
    
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
```
